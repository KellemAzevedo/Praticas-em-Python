```python
#Criando uma biblioteca com já alguns arquivos salvos.
funcionarios = {1: {'nome': 'Matheus', 'cargo': 'Consultor', 'cidade': 'Rj', 'estado': 'Rj', 'idade': 20, 'escolaridade': 'Superior', 'salario': 1000.0}, 2: {'nome': 'Kellen', 'cargo': 'Analista', 'cidade': 'Sp', 'estado': 'Sp', 'idade': 20, 'escolaridade': 'Superior', 'salario': 1000.0}} # Criando um dicionário vazio


#Criando uma Def para cadastrar um novo funcionário à biblioteca
def adicionar_funcionarios():
  while True:
    try:
      id = int(input('Digite o ID do funcionário: '))
      if id in funcionarios:
        print('esse id ja existe tente outro')
      else:
        break
    except:
      print('o valor informado não é um número')

  nome = input('Digite o nome do funcionário: ')
  cargo = input('Digite o cargo do funcionário: ')
  cidade = input('Digite a cidade do funcionário: ')
  estado = input('Digite o estado do funcionário: ')
  while True:
    try:
       idade = int(input('Digite a idade do funcionário: '))
       break
    except:
      print('o valor informado não é um número')

  escolaridade = input('Digite a escolaridade do funcionário: ')

  while True:
    try:
       salario = float(input('Digite o salário do funcionário: '))
       break
    except:
      print('o valor informado não é um número')

  funcionarios[id] = {'nome': nome, 'cargo': cargo, 'cidade': cidade, 'estado': estado, 'idade': idade, 'escolaridade': escolaridade, 'salario': salario}


#--------------------------------------------------------------------------------

#Criando uma Def para atualizar o cadastro de funcionário existente na biblioteca
def atualizar_funcionarios():
  while True:
    try:
       id = int(input('Digite o id do funcionário que deseja atualizar: '))
       break
    except:
      print('o valor informado não é um número')


  if id in funcionarios:
    print('Funcionário encontrado')
    nome = input('Digite o nome do funcionário: ')
    cargo = input('Digite o cargo do funcionário: ')
    cidade = input('Digite a cidade do funcionário: ')
    estado = input('Digite o estado do funcionário: ')
    while True:
     try:
       idade = int(input('Digite a idade do funcionário: '))
       break
     except:
      print('o valor informado não é um número')
    escolaridade = input('Digite a escolarida do funcionário: ')
    while True:
     try:
       salario = float(input('Digite o salário do funcionário: '))
       break
     except:
      print('o valor informado não é um número')

    funcionarios[id] = {'nome': nome, 'cargo': cargo, 'cidade': cidade, 'estado': estado, 'idade': idade, 'escolaridade': escolaridade, 'salario': salario}
  else:
    print('--------Usuário não encontrado-----------')

#---------------------------------------------------------------------------------

#Criando uma Def para exibir os dados de um funcionário específico existente na biblioteca
def exibir_funcionario():
  while True:
    try:
       id = int(input('Digite o id do funcionário que deseja exibir: '))
       break
    except:
      print('o valor informado não é um número')

  if id in funcionarios:
      d = funcionarios[id]
      print("ID:", id)
      print("Nome:", d['nome'])
      print("Cargo:", d['cargo'])
      print("Cidade:", d['cidade'])
      print("Estado:", d['estado'])
      print("idade:", d['idade'])
      print("Escolaridade:", d['escolaridade'])
      print("Salário:", d['salario'])
      print('_____________________________________')
  else:
    print('-------Funcionário não encontrado--------')
#---------------------------------------------------------------------------------

#Criando uma Def para listar todos os funcionários existentes na biblioteca
def listar_funcionarios(): #
  if funcionarios:
    for id, d in funcionarios.items():
      print("ID:", id)
      print("Nome:", d['nome'])
      print("Cargo:", d['cargo'])
      print("Cidade:", d['cidade'])
      print("Estado:", d['estado'])
      print("idade:", d['idade'])
      print("Escolaridade:", d['escolaridade'])
      print("Salário:", d['salario'])
      print('_________________________________')
  else:
    print('------Nenhum funcionário cadastrado-------')

#---------------------------------------------------------------------------------

#Criando uma Def para deletar um funcionário da biblioteca
def del_funcionarios():
   id = int(input('Digite o id do funcionário que deseja DELETAR: '))
   if id in funcionarios:
        del funcionarios[id]
   else:
      print('-------Funcionário não encontrado--------')

#--------------------------------------------------------------------------------

#estrutura de repetição para executar o programa até que o usuário insira a opção 6 (sair)
while True:
   print('=================Sistemas_De_Cadastro====================')
   print('Adicionar funcionário: Digite 1 ')
   print('Atualizar funcionário: Digite 2 ')
   print('Exibir funcionário: Digite 3 ')
   print('Listagem de funcionários: Digite 4')
   print('Excluir funcionário: Digite 5')
   print('Se você deseja sair: Digite 6')

   D = input('escolha uma opção: ')


   if D == '1':
      adicionar_funcionarios()
   elif D == '2':
      atualizar_funcionarios()
   elif D == '3':
      exibir_funcionario()
   elif D == '4':
      listar_funcionarios()
   elif D == '5':
      del_funcionarios()
   elif D == '6':
     print('Saindo do programa....')
     break
   else:
    print(f'Valor {D} invalido Digite (1, 2, 3, 4, 5 ou 6)')
```
